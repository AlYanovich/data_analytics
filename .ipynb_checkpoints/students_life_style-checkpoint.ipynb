Some exploratory analysis of all columns in dataset
Try to visualise distribution with histplot and boxplot

```python
plt.scatter(df.Study_Hours_Per_Day, df.GPA)
plt.title('Study hours vs GPA')
plt.xlabel('Study hours')
plt.ylabel('GPA')
plt.show()
```

There is positive dependence between Study_Hours and GPA. More hours of studing -> greater GPA student achive

```python
sns.scatterplot(data=df, x='Study_Hours_Per_Day', y='GPA', hue='Stress_Level', hue_order=['Low', 'Moderate', 'High'], alpha=0.7)
plt.show()
```

More hours of studing -> greater GPA student achive, but its lead to higher level of stress between students, but high level of strees observed all over all graph

```python
plt.figure(figsize=(8,4))

plt.ylabel('Count', fontsize=16)
plt.xlabel('Study hours per day', fontsize=16)
plt.title('Study hours per day distribution', fontsize=20)
plt.xticks(rotation=90)

sns.histplot(data=df, x=df['Study_Hours_Per_Day'], kde=True)
plt.show()
```

```python
plt.figure(figsize=(8,4))

plt.ylabel('Count', fontsize=16)
plt.xlabel('Study hours per day', fontsize=16)
plt.title('Study hours per day distribution', fontsize=20)
plt.xticks(rotation=90)
#plt.style.use('default') #ty to find some goodlooking styles and colors
#sns.color_palette("default")
order=['Low','Moderate','High']

sns.histplot(data=df, x=df['Study_Hours_Per_Day'], kde=True, hue='Stress_Level', hue_order=order, alpha=0.7)
plt.show()
```
