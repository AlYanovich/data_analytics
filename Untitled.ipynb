```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

what promotes or not promotes students to achive greate GPA and how it influence on their stress level?

```python
df = pd.read_csv('student_lifestyle_dataset.csv')
df.head()
```

```python
#general info about dataset
df.info()
```

```python
#statistical info about dataset
df.describe()
```

```python
#find any missing values but have the same info in df.info().['Non-Null Count']
df.isna().sum()
```

```python
df.columns
```

```python
plt.scatter(df.Study_Hours_Per_Day, df.GPA)
plt.title('Study hours vs GPA')
plt.xlabel('Study hours')
plt.ylabel('GPA')
plt.show()
```

There is positive dependence between Study_Hours and GPA. More hours of studing -> greater GPA student achive

```python
sns.scatterplot(data=df, x='Study_Hours_Per_Day', y='GPA', hue='Stress_Level', hue_order=['Low', 'Moderate', 'High'], alpha=0.5)
plt.show()
```

More hours of studing -> greater GPA student achive, but its lead to higher level of stress between students

```python
plt.scatter(df.Extracurricular_Hours_Per_Day, df.GPA)
plt.title('Extracurricular hours vs GPA')
plt.xlabel('Extracurricular hours')
plt.ylabel('GPA')
plt.show()
```

There is positive dependence between Sleep_Hours and GPA. More hours of sleeping -> greater GPA student achive

```python
sns.scatterplot(data=df, x='Extracurricular_Hours_Per_Day', y='GPA', hue='Stress_Level', hue_order=['Low', 'Moderate', 'High'], alpha=0.7)
plt.show()
```
